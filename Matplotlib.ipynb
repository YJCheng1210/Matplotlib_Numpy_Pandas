Draw a Sin Wave

```python
import matplotlib.pyplot as plt 
import numpy as np 

# Calculate the x, y axis of a sin curve 
x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x)
plt.title("sine wave form")  
plt.plot(x, y) 
plt.show()
```

# subplot()
subplot() function allows to plot different curves in the same figure.
The following example plots sine and cosine values:

```python
import numpy as np 
import matplotlib.pyplot as plt 
# Calculate the x and y axis of the sine and cosine curves. 
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x) 
y_cos = np.cos(x)  
# build a subplot frame, height=2, width=1  
# activate the subplot
plt.subplot(2,  1,  1)  
# draw the first figure
plt.plot(x, y_sin) 
plt.title('Sine')  
# activate the second subplot and draw the second figure
plt.subplot(2,  1,  2) 
plt.plot(x, y_cos) 
plt.title('Cosine')  
# show it
plt.show()
```
